```python
from PyPDF2 import PdfReader,PdfFileWriter

reader = PdfReader("MATHEMATICAL ASSOCIATION (2555-2560).pdf")


pages_section = [7, 13, 33, 40, 62, 70, 88, 96, 115, 124, 146, 154, 176] #page
name_section = ["math_assoctation(55-60)_sect-","math_assoctation(55-60)_(key)_sect-"]
# name_section = ["math_assoctation_sect-","math_assoctation_(key)_sect-"] 
n_section = 1
n_key_section = 1
```

```python
for j in range(len(pages_section)-1):
      pdfWriter = PdfFileWriter()
      for x in range(pages_section[j]-3,pages_section[j+1]-3):
            print(x)
            page = reader.getPage(x)
            # page.cropBox.setLowerLeft((40,20))
            # page.cropBox.setUpperRight((476,679))
            pdfWriter.addPage(page)

      if j%2 == 0:
         print("section")
         with open(name_section[0]+str(n_section)+".pdf", "wb") as output_stream:
            pdfWriter.write(output_stream)
         n_section = n_section+1
         del pdfWriter
      else :
         print("Key")
         with open(name_section[1]+str(n_key_section)+".pdf", "wb") as output_stream:
            pdfWriter.write(output_stream)
         n_key_section = n_key_section+1
         del pdfWriter  
```
